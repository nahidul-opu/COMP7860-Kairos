```python
import functools
import os
import json
import re
import torch
from tqdm import tqdm
from torch_geometric.data import *
import networkx as nx
import hashlib

def stringtomd5(originstr):
    """
    Convert a given string to its MD5 hash representation.

    Parameters:
    originstr (str): The original string to be hashed.

    Returns:
    str: The MD5 hash of the input string in hexadecimal format (64 characters).
    """
    originstr = originstr.encode("utf-8")
    signaturemd5 = hashlib.sha256()
    signaturemd5.update(originstr)
    return signaturemd5.hexdigest()

# path of folder that containes the JSON files of Clearscope e3 dataset
filePath="/home/cs/grad/opumni/Fall2024/COMP7860_Project/DatasetE3/data/clearscope/"

filelist = os.listdir(filePath)
```

```python
from datetime import datetime, timezone
import time
import pytz
from time import mktime
from datetime import datetime
import time

def ns_time_to_datetime(ns):
    """
    :param ns: int nano timestamp
    :return: datetime   format: 2013-10-10 23:40:00.000000000
    """
    dt = datetime.fromtimestamp(int(ns) // 1000000000)
    s = dt.strftime('%Y-%m-%d %H:%M:%S')
    s += '.' + str(int(int(ns) % 1000000000)).zfill(9)
    return s

def ns_time_to_datetime_US(ns):
    """
    :param ns: int nano timestamp
    :return: datetime   format: 2013-10-10 23:40:00.000000000
    """
    tz = pytz.timezone('US/Eastern')
    dt = pytz.datetime.datetime.fromtimestamp(int(ns) // 1000000000, tz)
    s = dt.strftime('%Y-%m-%d %H:%M:%S')
    s += '.' + str(int(int(ns) % 1000000000)).zfill(9)
    return s

def time_to_datetime_US(s):
    """
    :param ns: int nano timestamp
    :return: datetime   format: 2013-10-10 23:40:00
    """
    tz = pytz.timezone('US/Eastern')
    dt = pytz.datetime.datetime.fromtimestamp(int(s), tz)
    s = dt.strftime('%Y-%m-%d %H:%M:%S')

    return s

def datetime_to_ns_time(date):
    """
    :param date: str   format: %Y-%m-%d %H:%M:%S   e.g. 2013-10-10 23:40:00
    :return: nano timestamp
    """
    timeArray = time.strptime(date, "%Y-%m-%d %H:%M:%S")
    timeStamp = int(time.mktime(timeArray))
    timeStamp = timeStamp * 1000000000
    return timeStamp

def datetime_to_ns_time_US(date):
    """
    :param date: str   format: %Y-%m-%d %H:%M:%S   e.g. 2013-10-10 23:40:00
    :return: nano timestamp
    """
    tz = pytz.timezone('US/Eastern')
    timeArray = time.strptime(date, "%Y-%m-%d %H:%M:%S")
    dt = datetime.fromtimestamp(mktime(timeArray))
    timestamp = tz.localize(dt)
    timestamp = timestamp.timestamp()
    timeStamp = timestamp * 1000000000
    return int(timeStamp)

def datetime_to_timestamp_US(date):
    """
    :param date: str   format: %Y-%m-%d %H:%M:%S   e.g. 2013-10-10 23:40:00
    :return: nano timestamp
    """
    tz = pytz.timezone('US/Eastern')
    timeArray = time.strptime(date, "%Y-%m-%d %H:%M:%S")
    dt = datetime.fromtimestamp(mktime(timeArray))
    timestamp = tz.localize(dt)
    timestamp = timestamp.timestamp()
    timeStamp = timestamp
    return int(timeStamp)
```

## Netflow

Iterate over all the json files, collect the netflow related inforamtion.
Steps:
1. Find the Node ID (UUID), source IP address and port, and destination IP address and port
2. Concatnate ip and port to create  property string: `srcIP + "," srcPort + "," + destId + "," + destPort`
3. Create hash of the property string
4. Store the NodeID(UUID), hash and property string in a map

```python
# contains hash of each nodes property
#netobjset=set()

# netobj2has contains 2 types of data, a list for the NodeID as Key and NodeId for the Hash as Key. 
# netobj2hash[nodeid] = [hash(`srcIP + "," srcPort + "," + destId + "," + destPort`), `srcIP + "," srcPort + "," + destId + "," + destPort`]
# netobj2hash[hash(`srcIP + "," srcPort + "," + destId + "," + destPort`)] = nodeid
netobj2hash={}

for file in tqdm(filelist):
        with open(filePath + file, "r") as f:
            for line in f:
                if "NetFlowObject" in line:
                    try:
                        # find all returns list of all matches
                        res=re.findall('NetFlowObject":{"uuid":"(.*?)"(.*?)"localAddress":"(.*?)","localPort":(.*?),"remoteAddress":"(.*?)","remotePort":(.*?),',line)[0]

                        nodeid=res[0]
                        srcaddr=res[2]
                        srcport=res[3]
                        dstaddr=res[4]
                        dstport=res[5]

                        nodeproperty=srcaddr+","+srcport+","+dstaddr+","+dstport 
                        hashstr=stringtomd5(nodeproperty)
                        netobj2hash[nodeid]=[hashstr,nodeproperty]
                        netobj2hash[hashstr]=nodeid
                        #netobjset.add(hashstr)
                    except:
                        pass

```

```python
# Contains the data to be inserted into Postgres netflow_node_table
datalist=[]

for i in netobj2hash.keys():
    # check if the key represents a NodeID. The hash as length 64 and NodeID has langth 36 (standard uuid length)
    if len(i)!=64:
        datalist.append([i]+[netobj2hash[i][0]]+netobj2hash[i][1].split(","))


# insert into db table
sql = '''insert into netflow_node_table
                     values %s
        '''
ex.execute_values(cur,sql, datalist,page_size=10000)
connect.commit()  
```

```python
from sklearn.feature_extraction import FeatureHasher
from torch_geometric.transforms import NormalizeFeatures

from sklearn import preprocessing
import numpy as np

FH_string=FeatureHasher(n_features=16,input_type="string")
FH_dict=FeatureHasher(n_features=16,input_type="dict")



def path2higlist(p):
    """
    Generate a list of hierarchical paths from a given file path.

    Parameters:
    p (str): A string representing the file path.

    Returns:
    list: A list of hierarchical paths, starting from the base and building up.

    Example:
    >>> path2higlist("/home/user/docs/file.txt")
    ['home', 'home/user', 'home/user/docs', 'home/user/docs/file.txt']
    """
    l=[]
    spl=p.strip().split('/')
    for i in spl:
        if len(l)!=0:
            l.append(l[-1]+'/'+i)
        else:
            l.append(i)
    return l

def ip2higlist(p):
    """
    Generate a list of hierarchical segments from an IP address.

    Parameters:
    p (str): The IP address as a string.

    Returns:
    list: A list of hierarchical segments, starting from the first octet.

    Example:
    >>> ip2higlist("192.168.1.1")
    ['192', '192.168', '192.168.1', '192.168.1.1']
    """
    l=[]
    spl=p.strip().split('.')
    for i in spl:
        if len(l)!=0:
            l.append(l[-1]+'.'+i)
        else:
            l.append(i)
    return l


def subject2higlist(p):
    """
    Generate a list of hierarchical segments from a process name.

    Parameters:
    process_name (str): The name of the process (e.g., "python.main.module").

    Returns:
    list: A list of hierarchical segments, building from the first component.

    Example:
    >>> subject2higlist("python.main.module")
    ['python', 'python.main', 'python.main.module']
    """
    l=[]
    spl=p.strip().split('.')
    for i in spl:
        if len(l)!=0:
            l.append(l[-1]+'.'+i)
        else:
            l.append(i)
    return l


def list2str(l):
    """
    Convert a list of strings into a single concatenated string.

    Parameters:
    l (list): A list of strings.

    Returns:
    str: A single concatenated string containing all elements of the list.

    Example:
    >>> list2str(["Hello", " ", "World", "!"])
    'Hello World!'
    """
    s=''
    for i in l:
        s+=i
    return s
```

```python
node_msg_vec=[]
node_msg_dic_list=[]

# nodeid2msg has two types of keys 
# 1. nodeid2msg[hash] = index (int)
# 2. nodeid2msg[index] = dict{type:msg}
for i in tqdm(nodeid2msg.keys()):

    # check if the element is of 2nd type
    if type(i)==int:

        # check if the type of `dict{type:msg}` is netflow and create heirarchical feature list
        if 'netflow' in nodeid2msg[i].keys():
            higlist=['netflow']
            higlist+=ip2higlist(nodeid2msg[i]['netflow'])
        
        # check if the type of `dict{type:msg}` is file and create heirarchical feature list    
        if 'file' in nodeid2msg[i].keys():
            higlist=['file']
            higlist+=path2higlist(nodeid2msg[i]['file'])

        # check if the type of `dict{type:msg}` is subject and create heirarchical feature list
        if 'subject' in nodeid2msg[i].keys():
            higlist=['subject']
            higlist+=subject2higlist(nodeid2msg[i]['subject'])
            
        node_msg_dic_list.append(list2str(higlist))
```

```python
# Hash heirarchical feature list to create feature vector
node2higvec=[]
for i in tqdm(node_msg_dic_list):
    vec=FH_string.transform([i]).toarray()
    node2higvec.append(vec)
```

```python
node2higvec=np.array(node2higvec).reshape([-1,16])
```
